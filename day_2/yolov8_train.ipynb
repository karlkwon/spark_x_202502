```python
model2 = YOLO('yolov8s_trained.pt')
```

```python
def inference3(path):
    img_color = cv2.imread(path)
    results = model2(img_color)
    plots = results[0].plot()
    plt.imshow(cv2.cvtColor(plots, cv2.COLOR_BGR2RGB))
    plt.show()

    return results
```

```python
def convert_box_center_2_mm(results, img_size, pix_2_mm):
    cls = results[0].boxes.cls.tolist()
    centers = []
    
    for i, r in enumerate(results[0].boxes.xywh):
        d = r.tolist()
        if info[0].boxes.conf[i] > 0.8:
            centers.append( (cls[i], (d[0] - img_size[0]/2)*pix_2_mm, (d[1] - img_size[1]/2)*pix_2_mm) )
    
    return centers
```

```python
pix_2_mm = 0.12
```

```python
convert_box_center_2_mm(info, (1280, 720), pix_2_mm)
```
